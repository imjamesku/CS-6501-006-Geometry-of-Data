## 1 Geodesic Equation on $S^2$
### (a)
$v^{\theta}=\frac{\partial s}{\partial \theta} = (-\sin\theta \cos \phi, \cos \theta \cos \phi, 0)$

$v^{\phi}=\frac{\partial s}{\partial \phi} = (-\cos \theta \sin \phi, -\sin \theta \sin \phi, \cos \phi)$

### (b)
$g_{ij} = \langle v^i, v^j\rangle$, so

$$g = \begin{bmatrix}
    \sin^2\theta \cos^2\phi + \cos^2\theta \cos^2\phi & \sin\theta\cos\theta\sin\phi\cos\phi - \sin\theta\cos\theta\sin\phi\cos\phi\\
    \sin\theta\cos\theta\sin\phi\cos\phi - \sin\theta\cos\theta\sin\phi\cos\phi  & \cos^2\theta\sin^2\phi + \sin^2\theta\sin^2\phi+\cos^2\theta
\end{bmatrix} =
\begin{bmatrix}
    \cos^2\phi(\sin^2\theta+\cos^2\theta) & 0\\
    0  & \sin^2\phi(\cos^2\theta+\sin^2\theta) + \cos^2\phi
\end{bmatrix} = 
\begin{bmatrix}
    \cos^2\phi & 0\\
    0  & 1
\end{bmatrix}
$$

### (c)

$$g^{-1}=\frac{1}{\cos^2\phi}\begin{bmatrix}
    1 & 0\\
    0  & \cos^2\phi
\end{bmatrix}=
\begin{bmatrix}
    \frac{1}{\cos^2\phi} & 0\\
    0  & 1
\end{bmatrix}
$$

### (d)

When $k=\theta$

$$\Gamma^{\theta}_{ij}=\frac{1}{2}g^{\theta l}(\frac{\partial g_{jl}}{\partial x^i} + \frac{\partial g_{il}}{\partial x^j} - \frac{\partial g_{ij}}{\partial x^l})$$

There are 4 Christoffel symbols when $k=\phi$

$$\Gamma^\theta_{\theta\theta} = \frac{1}{2}\frac{1}{\cos^2\phi}(\frac{\partial g_{\theta\theta}}{\partial x^\theta} + \frac{\partial g_{\theta\theta}}{\partial x^\theta} - \frac{\partial g_{\theta\theta}}{\partial x^\theta}) = 0$$

$$\Gamma^\theta_{\theta\phi} = \frac{1}{2}\frac{1}{\cos^2\phi}(\frac{\partial g_{\phi\theta}}{\partial x^\theta} + \frac{\partial g_{\theta\theta}}{\partial X^\phi} - \frac{\partial g_{\theta\phi}}{\partial x^\theta})=\frac{1}{2\cos^2\phi}(-2\cos\phi\sin\phi) = -\sin\phi\cos\phi$$

$$\Gamma^\theta_{\phi\theta} = \frac{1}{2}\frac{1}{\cos^2\phi}(\frac{\partial g_{\theta\theta}}{\partial x^\phi} + \frac{\partial g_{\phi\theta}}{x^\theta} - \frac{\partial g_{\phi\theta}}{\partial x^\theta})=\frac{1}{2\cos^2\phi}(-2\cos\phi\sin\phi) = -\sin\phi\cos\phi$$

$$\Gamma^\theta_{\phi\phi} = \frac{1}{2}\frac{1}{\cos^2\phi}(\frac{\partial g_{\phi\theta}}{\partial x^\phi} + \frac{\partial g_{\phi\theta}}{x^\phi} - \frac{\partial g_{\phi\phi}}{\partial x^\theta})=0$$

When $k=\phi$

$$\Gamma^{\phi}_{ij}=\frac{1}{2}g^{\phi \phi}(\frac{\partial g_{j\phi}}{\partial x^i} + \frac{\partial g_{i\phi}}{\partial x^j} - \frac{\partial g_{ij}}{\partial x^\phi})$$

There are 4 Christoffel symbols when $k=\phi$

$$\Gamma^\phi_{\theta\theta} = \frac{1}{2}(\frac{\partial g_{\theta\phi}}{\partial x^\theta} + \frac{\partial g_{\theta\phi}}{\partial x^\theta} - \frac{\partial g_{\theta\theta}}{\partial x^\theta}) = \frac{1}{2}(-2\cos\phi\sin\phi) = -\cos\phi\sin\phi$$

$$\Gamma^\phi_{\theta\phi} = \frac{1}{2}(\frac{\partial g_{\phi\phi}}{\partial x^\theta} + \frac{\partial g_{\theta\phi}}{\partial x^\phi} - \frac{\partial g_{\theta\phi}}{\partial x^\phi})=0$$

$$\Gamma^\phi_{\phi\theta} = \frac{1}{2}(\frac{\partial g_{\theta\phi}}{\partial x^\phi} + \frac{\partial g_{\phi\phi}}{\partial x^\theta} - \frac{\partial g_{\phi\theta}}{\partial x^\phi})=0$$

$$\Gamma^\phi_{\phi\phi} = \frac{1}{2}(\frac{\partial g_{\phi\phi}}{\partial x^\phi} + \frac{\partial g_{\phi\phi}}{\partial x^\phi} - \frac{\partial g_{\phi\phi}}{\partial x^\phi})=0$$

### (e)
$$\frac{d^2s^1}{dt^2} = -\sum_{i,j=1}^{n}\Gamma_{ij}^1\frac{d\gamma^i}{dt}\frac{d\gamma^j}{dt} = -(-\sin\phi\cos\phi\frac{d\gamma^1}{dt}\frac{d\gamma^2}{dt} - \sin\phi\cos\phi\frac{d\gamma^2}{dt}\frac{d\gamma^1}{dt}) = 2\sin\phi\cos\phi\frac{d\gamma^1}{dt}\frac{d\gamma^2}{dt}$$

$$\frac{d^2s^2}{dt^2} = -(-\cos\phi\sin\phi \frac{d\gamma^1}{dt}\frac{d\gamma^1}{dt}) = \cos\phi\sin\phi\frac{d\gamma^1}{dt}\frac{d\gamma^1}{dt}$$
Since we are only considering the equator $s^1 = \theta = kt$, where k is the speed, and $s^2=\phi=0$, 

Substituting in the values of $s^1$ and $s^2$, we get $0=0$ and $0=0$ from both equations. Hence the euqator is a geodesic


### (f)
Since we have spherical symmetry, the equator being a geodesic means all great circles must also be geodesics.

## Statistics on Shape Manifolds

```python
import numpy as np
import matplotlib
import matplotlib.pyplot as plt

class Shape:
    def __init__(self, points):
        self.matrix = np.array(points).T
    def setMatrix(self, matrix):
        self.matrix = matrix
    def plot(self):
        x = [xi for xi in self.matrix[0, :]]
        y = [yi for yi in self.matrix[1, :]]
        plt.scatter(x, y)
#         plt.show()
    def project(self):
        # Project an onto the preshape sphere by removing its centroid and scaling it to unit norm
        for i in range(self.matrix.shape[0]):
            self.matrix[i] = self.matrix[i] - np.mean(self.matrix[i])
        self.matrix /= np.linalg.norm(self.matrix)
    def opa(self, target):
        # input a target preshape and moving preshape and output the aligned version of the moving preshape.
        u, s, vh = np.linalg.svd(target.matrix.dot(self.matrix.T), full_matrices=True)
        self.matrix = u.dot(vh.dot(self.matrix))
    def expMap(self, tangentVector):
        p = np.array(self.matrix)
        p += tangentVector
        p /= np.linalg.norm(p)
        endPoint = Shape([])
        endPoint.setMatrix(p)
        return endPoint
    def logMap(self, endPoint):
        return (endPoint.matrix -
                (self.matrix[0].dot(endPoint.matrix[0]) + self.matrix[1].dot(endPoint.matrix[1])) * self.matrix)*\
                (self.matrix[0].dot(self.matrix[0]) + self.matrix[1].dot(self.matrix[1]))/\
                (self.matrix[0].dot(endPoint.matrix[0]) + self.matrix[1].dot(endPoint.matrix[1]))

def noise():
    return np.random.normal(0, 0.1)
```

```python
import copy
def frechetMean(shapes, nIter):
    u = copy.deepcopy(shapes[0])
    for i in range(nIter):
        gradient = u.matrix - u.matrix
        for shape in shapes:
            gradient += u.logMap(shape)
        gradient /= len(shapes)
        u = u.expMap(gradient)
    return u
def pca(X):
    """returns eigenvectors and eigenvalues(columns)"""
    X = X - np.mean(X, 0)
    sigma = X.T.dot(X)/X.shape[0]
    return np.linalg.eig(sigma)
```

# Report
## My Implementation
1. `project()`: Subtracts the mean from every point and devide the matrix by its norm
2. `opa(self, target)`: $M = BA^T$. Uses singular value decomposition(which will give us $M = U\Sigma V^T$) to find $U$ and $V^T$ and return $R = UV^T$
3. `expMap(self, tangentVector)`: Adds the tangent vector to the preshape and divide the result by its norm
4. `logMap(self, endPoint)`: Calculates the tangent vector with $(y-(x\cdot y)x)\frac{|x|}{x\cdot y}$
5. `frechetMean`: Uses the first shape as the initial value. In each epoch, calculates the gradient for each shape and adds their sum to the current mean. Repeats `nIter` times.
6. Subtract the mean from all rows and then perform eigen decomposition on the covariance matrix

## Triangle Experiment

```python
# Generate points
triangles = [Shape([
    [-1 + noise(), 0 + noise()],
    [1 + noise(), 0 + noise()],
    [0 + noise(), np.random.normal(1, 0.5) + noise()]
]) for i in range(100)]
# print(triangles[0].matrix)
# triangles[0].plot()
# plt.show()
# triangles[0].project()
# triangles[0].plot()
# plt.show()
# print(triangles[0].matrix)
# print(np.linalg.norm(triangles[0].matrix))
for i in range(len(triangles)):
    triangles[i].project()
#     triangles[i].opa(triangles[0])
mean = frechetMean(triangles, 10)
mean.plot()
plt.show()
print(mean.matrix)
```

### Geodesic between 2 Objects

```python
triangles[0].plot()
axes = plt.gca()
axes.set_xlim([-1,1])
axes.set_ylim([-1,1])
plt.show()
start = copy.deepcopy(triangles[0])
end = triangles[34]
for i in range(5):
    tangent = start.logMap(end) * 0.2
    start = start.expMap(tangent)
    start.plot()
    axes = plt.gca()
    axes.set_xlim([-1,1])
    axes.set_ylim([-1,1])
#     plt.show()
plt.show()
end.plot()
axes = plt.gca()
axes.set_xlim([-1,1])
axes.set_ylim([-1,1])
plt.show()
```

Each step I use the log map to get a tangent vector between the 2 shapes and then add 0.3 of that tagent vector(if we just added the entire tangent vector, we would get to the end of the curve in 1 step) to the current point and then take the exponential map to get to the next step on the curve.
As the point moved along the geodesic, the shape got more similar to the second object

### Alignment Experiement

```python
# plot all objects
for triangle in triangles:
    triangle.plot()
axes = plt.gca()
axes.set_xlim([-1,1])
axes.set_ylim([-1,1])
plt.title('Raw Object Data')
plt.show()

mean = frechetMean(triangles, 10)

## plot aligned objects
alignedTriangles = copy.deepcopy(triangles)
for triangle in alignedTriangles:
    triangle.opa(mean)
    triangle.plot()
axes = plt.gca()
axes.set_xlim([-1,1])
axes.set_ylim([-1,1])
plt.title('Aligned Object Data')
plt.show()

mean.plot()
axes = plt.gca()
axes.set_xlim([-1,1])
axes.set_ylim([-1,1])
plt.title('Frechet Mean')
plt.show()
```

The difference between before alignment and after alignment is quite obvious and noticable, but not very big because the triangles weren't generated with a huge variance in orentation in the first place

### Plot of Eigenvalues

```python
# print(triangles[0].matrix)
# print(triangles[0].matrix.flatten())
flattenedTriangles = []
for triangle in triangles:
    flattenedTriangles.append(triangle.matrix.flatten())
flattenedTriangles = np.array(flattenedTriangles)
# print(flattenedTriangles.shape)
# print(flattenedTriangles)
eigenvalues, eigenvectors = pca(flattenedTriangles)
sortedValueVectorPairs = sorted([(eigenvalues[i], eigenvectors[:, i]) for i in range(len(eigenvalues))], reverse=True)
sortedEigenvalues = [pair[0] for pair in sortedValueVectorPairs]
sortedEigenVectors = [pair[1] for pair in sortedValueVectorPairs]
print(eigenvalues)
print(eigenvectors)
sing_vals = np.arange(eigenvalues.shape[0]) + 1
plt.plot(sing_vals, sortedEigenvalues, 'ro-', linewidth=2)
plt.title('Scree Plot')
plt.xlabel('Principal Component')
plt.ylabel('Eigenvalue')
#I don't like the default legend so I typically make mine like below, e.g.
#with smaller fonts and a bit transparent so I do not cover up data, and make
#it moveable by the viewer in case upper-right is a bad place for it 
leg = plt.legend(['Eigenvalues from SVD'], loc='best', borderpad=0.3, 
                 shadow=False, prop=matplotlib.font_manager.FontProperties(size='small'),
                 markerscale=0.4)
leg.get_frame().set_alpha(0.4)
leg.draggable(state=True)
plt.show()
```

I would use only 1 mode, since from 2 onwards we don't gain much variance from adding more modes.
