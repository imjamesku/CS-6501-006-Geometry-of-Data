```python
import torch
import torch.nn.functional as F
from torch.utils.data import DataLoader
from torchvision.datasets import FashionMNIST
import torchvision.utils as vutils
import torchvision.transforms as transforms
import torch.nn as nn
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
class FullyConnectedAutoencoder(nn.Module):
    def __init__(self, out_size):
        super(FullyConnectedAutoencoder, self).__init__()

        self.encoder = nn.Sequential(
            nn.Linear(input_size, 256),
            nn.BatchNorm1d(256),
            nn.ELU(),
            nn.Linear(256, 128),
            nn.BatchNorm1d(128),
            nn.ELU(),
            nn.Linear(128, out_size),
            nn.ELU()
        )

        self.decoder = nn.Sequential(
            nn.Linear(out_size, 128),
            nn.BatchNorm1d(128),
            nn.ELU(),
            nn.Linear(128, 256),
            nn.BatchNorm1d(256),
            nn.ELU(),
            nn.Linear(256, input_size),
            nn.Sigmoid()
        )
        
    def forward(self, x):
        x = x.view(-1, input_size)
        z = self.encoder(x)
        y = self.decoder(z)
        y = y.view(-1, 1, width, height)
        return y
    def encode(self, x):
        x = x.view(-1, input_size)
        z = self.encoder(x)
        return z
    def decode(self, z):
        y = self.decoder(z)
        y = y.view(-1, 1, width, height)
        return y
class ConvolutionalAutoencoder(nn.Module):
    def __init__(self, out_size):
        super(ConvolutionalAutoencoder, self).__init__()
        self.econv1 = nn.Conv2d(1, 32, kernel_size = 5, bias = False)
        self.ebatch1 = nn.BatchNorm2d(32)
        self.econv2 = nn.Conv2d(32, 8, kernel_size = 5, bias = False)
        self.ebatch2 = nn.BatchNorm2d(8)
        self.econv3 = nn.Conv2d(8, out_size, kernel_size = 20, bias = True)

        self.dconv1 = nn.ConvTranspose2d(out_size, 8, kernel_size = 20, bias = True)
        self.dbatch1 = nn.BatchNorm2d(8)
        self.dconv2 = nn.ConvTranspose2d(8, 32, kernel_size = 5, bias = False)
        self.dbatch2 = nn.BatchNorm2d(32)
        self.dconv3 = nn.ConvTranspose2d(32, 1, kernel_size = 5, bias = False)

    def encoder(self, x):
        y1 = F.elu(self.ebatch1(self.econv1(x)))
        y2 = F.elu(self.ebatch2(self.econv2(y1)))
        z = F.elu(self.econv3(y2))
        return z

    def decoder(self, z):
        y1 = F.elu(self.dbatch1(self.dconv1(z)))
        y2 = F.elu(self.dbatch2(self.dconv2(y1)))
        x = torch.sigmoid(self.dconv3(y2))
        return x
    def encode(self, x):
        x = x.reshape(-1, 1, 28, 28)
        return self.encoder(x)
    def decode(self, z):
        return self.decoder(z)
    def forward(self, x):
        z = self.encoder(x)
        y = self.decoder(z)
        return y
```

## 3. PCA

```python
class PCA_CLASS:
    def init(self, X):
        self.fit(X)
    def fit(self, X):
        # preprocess the data
        X = X.reshape((-1, height*width))
        X_mean = torch.mean(X,0)
        self.mean = X_mean
        X = X - X_mean.expand_as(X)
        # svd
        U,S,V = torch.svd(X)
        self.U, self.S, self.V = U, torch.diag(S), V
#         print(self.U.shape)
#         print(U)
#         print(self.S.shape)
#         print(S)
#         print(self.V.shape)
    def encode(self, X, k):
        X = X.reshape((-1, height*width))
        X = X - self.mean.expand_as(X)
        Vk = self.V[:, :k]
        return X.mm(Vk)
    def decode(self, Z, k):
        Vk = self.V[:, :k]
        X = Z.mm(Vk.t())
        X = X + self.mean.expand_as(X)
        return X.reshape(-1, 1, 28, 28)
```

```python
pca = PCA_CLASS()
pca.fit(train_data_matrix)
```

```python
batch = first_batch[0]
k = 128
plot_images(batch, 10, 10, 'original')

for k in [16, 32, 128]:
    test_e = pca.encode(batch, k)
    test_d = pca.decode(test_e, k)
    plot_images(test_d, 10, 10, 'k={}'.format(k))
```

```python
def getPath(p1, p2, stops):
    path = []
    diff = p2 - p1
    step = diff/(stops+1)
    path.append(p1)
    curr = p1.clone()
    for i in range(stops):
        curr += step
        path.append(curr.clone())
    path.append(p2)
    return path
def interpolate(p1, p2, model):
    model.eval()
    encodedP1 = model.encode(p1)
    encodedP2 = model.encode(p2)
    path = getPath(encodedP1, encodedP2, 10)
    path = [model.decode(p)[0] for p in path]
    return path
```

```python
from torch.autograd.gradcheck import zero_gradients
from torch.autograd import Variable
# compute the Jacobian matrix of the decoder of a given model
def get_jacobian(net, x, noutputs):
#     x.requires_grad=True
    x = Variable(x, requires_grad=True)
#     x.grad.data.zero_()
    y = net(x)
    jacobian = torch.zeros(noutputs, *x.size())
#     print(jacobian.shape)
    grad_output = torch.zeros(*y.size())
#     print('grad output shape: {}'.format(grad_output_output.shape))
    if x.is_cuda:
        grad_output = grad_output.cuda()
        jacobian = jacobian.cuda()
    for i in range(noutputs):
        zero_gradients(x)
        grad_output.zero_()
        grad_output[:, i] = 1
        y.backward(grad_output, retain_graph=True)
        jacobian[i] = x.grad.data
    jacobian = jacobian.view(784, -1)
    return jacobian
```

### fcAE16

```python
model = fcAE16
p1.requires_grad=True
encodedP1 = model.encode(p1)
jacobian = get_jacobian(model.decoder, encodedP1, height*width)
print('jacobian:')
print(jacobian)
U, S, V = torch.svd(jacobian)
print('singular values:')
print(S)
print('rank of jacobian: {}'.format(np.linalg.matrix_rank(jacobian.numpy())))
```

The jacobian matrix is full rank, meaning the decoder is an immersion.

```python
def get_jacobian_conv(net, x, height, width):
#     x.requires_grad=True
    x = Variable(x, requires_grad=True)
#     x.grad.data.zero_()
    y = net(x)
    jacobian = torch.zeros(height*width, *x.size())
#     print(jacobian.shape)
    grad_output = torch.zeros(*y.size())
#     print('grad output shape: {}'.format(grad_output.shape))
    if x.is_cuda:
        grad_output = grad_output.cuda()
        jacobian = jacobian.cuda()
    for i in range(height):
        for j in range(width):
            zero_gradients(x)
            grad_output.zero_()
            grad_output[:, :, i, j] = 1
            y.backward(grad_output, retain_graph=True)
            jacobian[i*width+j] = x.grad.data
    jacobian = jacobian.view(784, -1)
    return jacobian
```

### fcAE128

```python
model = convAE16
p1.requires_grad=True
encodedP1 = model.encode(p1)
jacobian = get_jacobian_conv(model.decoder, encodedP1, height, width)
print('jacobian:')
print(jacobian)
U, S, V = torch.svd(jacobian)
# print(jacobian.shape)
print('singular values:')
print(S)
print('rank: {}'.format(np.linalg.matrix_rank(jacobian.numpy())))
```

Jocobian matrix is full rank, so the decoder of convAE16 is an immersion

```python
def plotProgressions(model, dims, base, steps, lim=5, stepsize = 1):
    model.eval()
    base.reshape(-1, 28, 28)
    with torch.no_grad():
        plot_images(base, 1, 1, 'Base Image')
    vectors = []
    for i in range(dims):
        vec = [0.0] * dims
        vec[i] = 1.0
        vectors.append(vec)
    for i in range(lim):
        encodedBase = model.encode(base)
        v = vectors[i]
        t = torch.tensor(v).reshape(encodedBase.shape)
        cur = encodedBase
        path = []
        for _ in range(steps):
            path.append(model.decode(cur))
            cur = cur + t * stepsize
        # print(decoded)
#         print(path[-1].shape)
        path = torch.stack(path)
        path = path.reshape(-1,1, 28, 28)
#         print(path.shape)
        with torch.no_grad():
            plot_images(path, steps, 1, str(v)[:32] + '...')
```

```python
def plotPCAProgressions(pca, k, base, steps, lim = 10):
    base.reshape(-1, 28, 28)
    with torch.no_grad():
        plot_images(base, 1, 1, 'Base Image')
    vectors = []
    for i in range(k):
        vec = [0.0] * k
        vec[i] = 1.0
        vectors.append(vec)
    encodedBase = pca.encode(base, k)
    print(encodedBase.shape)
    for i in range(lim):
        v = vectors[i]
        t = torch.tensor(v)
        cur = encodedBase.clone().detach()
        path = []
        for _ in range(10):
            path.append(pca.decode(cur, k).clone())
            cur = cur + t
        # print(decoded)
        path = torch.stack(path)
        path = path.reshape(-1,1, 28, 28)
#         print(path.shape)
        with torch.no_grad():
            plot_images(path, steps, 1, str(v)[:32] + '...')
```

```python
# translation
def translate(tensor, shift, axis):
    if shift == 0:
        return tensor

    if axis < 0:
        axis += tensor.dim()

    dim_size = tensor.size(axis)
    after_start = dim_size - shift
    if shift < 0:
        after_start = -shift
        shift = dim_size - abs(shift)

    before = tensor.narrow(axis, 0, dim_size - shift)
    after = tensor.narrow(axis, after_start, shift)
    return torch.cat([after, before], axis)

model = fcAE16
delt = 1
base = first_batch[0][1]
shifted = translate(base, delt, 2)
encodedBase = model.encode(base)
encodedShifted = model.encode(shifted)
tangent = (encodedShifted - encodedBase)/delt
print('tangent vector: {}'.format(tangent))
```

### Rotation

```python
def rotate_coords(x, y, theta, ox, oy):
    """Rotate arrays of coordinates x and y by theta radians about the
    point (ox, oy).

    """
    s, c = np.sin(theta), np.cos(theta)
    x, y = np.asarray(x) - ox, np.asarray(y) - oy
    return x * c - y * s + ox, x * s + y * c + oy
def rotate_image(src, theta, ox, oy, fill=255):
    theta = -theta
    sh, sw = src.shape
    cx, cy = rotate_coords([0, sw, sw, 0], [0, 0, sh, sh], theta, ox, oy)
    dw, dh = sw, sh
    # Coordinates of pixels in destination image.
    dx, dy = np.meshgrid(np.arange(dw), np.arange(dh))
    sx, sy = rotate_coords(dx + cx.min(), dy + cy.min(), -theta, ox, oy)
    sx, sy = sx.round().astype(int), sy.round().astype(int)
    mask = (0 <= sx) & (sx < sw) & (0 <= sy) & (sy < sh)
    dest = np.empty(shape=(dh, dw), dtype=src.dtype)
    dest[dy[mask], dx[mask]] = src[sy[mask], sx[mask]]
    dest[dy[~mask], dx[~mask]] = fill

    return dest
```
