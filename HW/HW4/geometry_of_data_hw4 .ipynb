## (a) Christofell symbols
$\Gamma_{ij}^k = \frac{1}{2}\sum_{l=1}^{n}g^{kl}(\frac{\partial g_{ij}}{\partial x^i} + \frac{\partial g_{il}}{\partial x^j} - \frac{\partial g_{ij}}{\partial x^l})$

$\Gamma_{\mu\mu}^{\mu}=\frac{1}{2}g^{\mu\mu}(\frac{\partial g_{\mu\mu}}{\partial\mu} + \frac{\partial g_{\mu\mu}}{\partial\mu} - \frac{\partial g_{\mu\mu}}{\partial\mu}) = \frac{1}{2}g^{\mu\mu}(0+0-0) = 0$

$\Gamma_{\mu\sigma}^{\mu}=\frac{1}{2}g^{\mu\mu}(\frac{\partial g_{\sigma\mu}}{\partial\mu} + \frac{\partial g_{\mu\mu}}{\partial\sigma} - \frac{\partial g_{\mu\sigma}}{\partial\mu}) = \frac{1}{2}g^{\mu\mu}(0-2\sigma^{-3}-0) = -\frac{1}{\sigma}$

$\Gamma_{\sigma\mu}^{\mu}=\frac{1}{2}g^{\mu\mu}(\frac{\partial g_{\mu\mu}}{\partial\sigma} + \frac{\partial g_{\sigma\mu}}{\partial\mu} - \frac{\partial g_{\sigma\mu}}{\partial\mu}) = \frac{1}{2}\frac{\sigma^4}{2}\frac{2}{\sigma^2}\cdot(-2)\sigma^{-3}=-\frac{1}{\sigma}$

$\Gamma_{\sigma\sigma}^{\mu}=\frac{1}{2}g^{\mu\mu}(\frac{\partial g_{\sigma\mu}}{\partial\sigma} + \frac{\partial g_{\sigma\mu}}{\partial\sigma} - \frac{\partial g_{\sigma\sigma}}{\partial\mu}) = \frac{1}{2}g^{\mu\mu}(0+0-0) = 0$

$\Gamma_{\mu\mu}^{\sigma}=\frac{1}{2}g^{\sigma\sigma}(\frac{\partial g_{\mu\sigma}}{\partial\mu} + \frac{\partial g_{\mu\sigma}}{\partial\mu} - \frac{\partial g_{\mu\mu}}{\partial\sigma}) = \frac{1}{2}\frac{\sigma^2}{2}\cdot 2\cdot \sigma^{-3} = \frac{1}{2\sigma}$

$\Gamma_{\mu\sigma}^{\sigma}=\frac{1}{2}g^{\sigma\sigma}(\frac{\partial g_{\sigma\sigma}}{\partial\mu} + \frac{\partial g_{\mu\sigma}}{\partial\sigma} - \frac{\partial g_{\mu\sigma}}{\partial\sigma}) = \frac{1}{2}g^{\sigma\sigma}(0+0-0) = 0$

$\Gamma_{\sigma\mu}^{\sigma}=\frac{1}{2}g^{\sigma\sigma}(\frac{\partial g_{\mu\sigma}}{\partial\sigma} + \frac{\partial g_{\sigma\sigma}}{\partial\sigma} - \frac{\partial g_{\sigma\mu}}{\partial\sigma}) = \frac{1}{2}g^{\sigma\sigma}(0+0-0) = 0$

$\Gamma_{\sigma\sigma}^{\sigma}=\frac{1}{2}g^{\sigma\sigma}(\frac{\partial g_{\sigma\sigma}}{\partial\sigma} + \frac{\partial g_{\sigma\sigma}}{\partial\sigma} - \frac{\partial g_{\sigma\sigma}}{\partial\sigma}) = \frac{1}{2}\frac{\sigma^2}{2}\cdot (-4)\cdot \sigma^{-3} = -\frac{1}{\sigma}$

## (b)
We need to show the geodesic equation: $\frac{d^2u^k}{dt^2} + \Gamma_{ij}^{k}\frac{du^i}{dt}\frac{du^j}{dt}$ is equal to $0$

For $k=\mu$,
$$\frac{d^2\mu}{dt^2} + \Gamma_{ij}^{\mu}\frac{du^i}{dt}\frac{du^j}{dt} = 0+2\frac{-1}{\sigma}\cdot 0 \cdot \frac{du^j}{dt} = 0$$

For $k=\sigma$
$$\frac{d^2\sigma}{dt^2} + \Gamma_{ij}^{\sigma}\frac{du^i}{dt}\frac{du^j}{dt} = bc^2e^{ct} + \frac{1}{2\sigma}\cdot 0 \cdot 0 + (-\frac{1}{\sigma})\frac{be^{ct}}{dt}\frac{be^{ct}}{dt}=bc^2e^{ct} + (-\frac{1}{be^{ct}})\cdot b^2c^2e^{2ct}=bc^2e^{ct}-bc^2e^{ct} = 0$$





## (c)
Again, we need to show the geodesic equation is euqual to 0.

For $k=\mu$
$$\frac{d^2a\tanh(ct)}{dt^2} + \Gamma^\mu_{ij}\frac{du^i}{dt}\frac{du^j}{dt}=2ac^2\frac{1}{\cosh(ct)}(-\frac{1}{\cosh(ct)}\tanh(ct)) + 2\frac{-\cosh(ct)}{a\cdot\sqrt{2}/2}\cdot ac\frac{1}{\cosh^2(ct)}\cdot a \frac{\sqrt{2}}{2}\cdot c\frac{-1}{\cosh(ct)}tanh(ct) = -2ac^2\frac{1}{\cosh^2(ct)}\tanh(ct) + 2ac^2\frac{1}{\cosh^2(ct)}\tanh(ct) = 0$$

For $k=\sigma$
$$\frac{d^2\frac{a\sqrt{2}}{2\cosh(ct)}}{dt^2} + \Gamma^\sigma_{ij}\frac{du^i}{dt}\frac{du^j}{dt} = -ac^2\frac{\sqrt{2}}{2}(-\frac{1}{\cosh(ct)} + \tanh^2(ct) + \frac{1}{\cosh^3(ct)}) - \frac{1}{\sigma}\frac{a^2}{2}(c\frac{-1}{\cosh}\tanh(ct))(\frac{-1}{\cosh(ct)}\tanh(ct)c) + \frac{1}{2\sigma}\cdot ac\frac{1}{\cosh^2(ct)} = -ac^2\frac{\sqrt{2}}{2}(-\frac{1}{\cosh(ct)} + \tanh^2(ct) + \frac{1}{\cosh^3(ct)}) + \frac{-ac^2\tanh^2(ct)}{\sqrt{2}\cosh(ct)} + \frac{ac^2}{\sqrt{2}\cosh^3(ct)} = 0$$

```
def plotGaussian(mu, sigma):
  x = np.linspace(mu - 3*sigma, mu + 3*sigma, 100)
  plt.plot(x, stats.norm.pdf(x, mu, sigma))
plotGaussian(0, 1)
plotGaussian(0, 2)
plt.show()

plt.plot([0,0],[1,2],'k-')
plt.plot([0,0],[1,2],'ro')
plt.show()
```

distance: $\int_1^2\begin{bmatrix}0 & 1\end{bmatrix}
\begin{bmatrix}
\frac{1}{t^2} & 0 \\
0             & \frac{2}{t^2}
\end{bmatrix}
\begin{bmatrix}0 \\
1\end{bmatrix}dt = \int_1^2\frac{2}{t^2}dt = 1$

```
plotGaussian(0, 4)
plotGaussian(0, 8)
plt.show()

plt.plot([0,0],[4,8],'k-')
plt.plot([0,0],[4,8],'ro')
plt.show()
```

distance: $\int_1^2\begin{bmatrix}0 & 4\end{bmatrix}

\end{bmatrix}
\begin{bmatrix}0 \\
4\end{bmatrix}dt = \int_1^2\frac{2}{t^2}dt = 1$

Scaling the standard deviation by the same vector results in the same geodesic distance regardless of the original $\mu$ and $\sigma$.

$$\int_{-1}^1

\begin{bmatrix}
\frac{1}{2-t^2} & 0 \\

\end{bmatrix}dt=\int_{-1}^1(\frac{1}{2-t^2}+\frac{2t}{(2-t^2)^2})dt = \int_{-1}^1\frac{2+t^2}{(2-t^2)^2}dt = 2$$

curve: $(2t, \sqrt{8-4t^2})$
distance:
$$\int_{-1}^1
\begin{bmatrix}
2 & \frac{-2t}{\sqrt{2-t^2}}
\end{bmatrix}
\begin{bmatrix}
\frac{1}{4(2-t^2)} & 0 \\
0               & \frac{2}{4(2-t^2)}
\end{bmatrix}
\begin{bmatrix}
2 \\
\frac{-2t}{\sqrt{2-t^2}}
\end{bmatrix}dt=\int_{-1}^1(\frac{1}{2-t^2}+\frac{2t}{(2-t^2)^2})dt = \int_{-1}^1\frac{2+t^2}{(2-t^2)^2}dt = 2$$

The distances turned out to be the same. This means the doubling of the parameters of two gaussian distributions does not change the distance between them.

```
class LogisticRegression:
  def __init__(self, initial_w, initial_b):
    self.w = np.array(initial_w)
    self.b = np.array(initial_b)
  def sigmoid(self, X):
    # print(X)
    return 1/(1 + np.exp(-X))
  def predict(self, X):
    return self.sigmoid(X @ self.w + self.b)
  def loss(self, predictions, y):
    return -np.sum(y@np.log(predictions) + (1-y)@np.log(1-predictions))/y.shape[0]
  def fit_GD(self, trainX, trainY, valX, valY, epoches, lr):
    n = trainX.shape[0]
    loss = []
    valLoss = []
    for _ in range(epoches):
      predictions = self.predict(trainX)
      gradient = np.zeros(self.w.shape[0])
      bias = 0
      for i, xi in enumerate(trainX):
        gradient += (trainY[i] - predictions[i]) * xi
        bias += (trainY[i] - predictions[i])
      self.w += lr * gradient/n
      self.b += lr * bias/n
      predictions = self.predict(trainX)
      loss.append(self.loss(predictions, trainY))
      valLoss.append(self.loss(self.predict(valX), valY))
    return loss, valLoss
      # print('w: {}'.format(self.w))
      # print('b: {}'.format(self.b))
  def fit_NGD(self, trainX, trainY, valX, valY, epoches, lr):
    # random initialization
    n = trainX.shape[0]
    loss = []
    valLoss = []
    for _ in range(epoches):
      predictions = self.predict(trainX)
      # print('predictions: {}'.format(predictions))
      gradient = np.zeros(self.w.shape[0]+1)
      bias = 0
      F = np.zeros((self.w.shape[0]+1, self.w.shape[0]+1))
      for i, xi in enumerate(trainX):
        xiGradient = (trainY[i] - predictions[i]) * np.concatenate(([1], xi))
        # print(xiGradient.shape)
        F += xiGradient.reshape(-1, 1) @ xiGradient.reshape(1, -1)
        gradient += xiGradient
      F /= n
      gradient /= n
      naturalGradient = np.linalg.pinv(F) @ gradient
      self.w += lr * naturalGradient[1:]
      self.b += lr * naturalGradient[0]
      predictions = self.predict(trainX)
      loss.append(self.loss(predictions, trainY))
      valLoss.append(self.loss(self.predict(valX), valY))
      # print('ngd: {}'.format(naturalGradient))
      # print('w: {}'.format(self.w))
      # print('b: {}'.format(self.b))
    return loss, valLoss


def accuracy(yHat, y):
  yClass = np.around(yHat)
  return y[yClass==y].shape[0]/y.shape[0]

```

## (a) Loading and Preprocessing Data

```
fashion_mnist = keras.datasets.fashion_mnist
(train_images, train_labels), (test_images, test_labels) = fashion_mnist.load_data()
```

```
train_images = train_images / 255.0
test_images = test_images / 255.0
train_images = train_images.reshape((-1, 28*28))

mask = np.logical_or(train_labels == 0, train_labels == 1)
train_images_subset = train_images[mask][:200]
train_labels_subset = train_labels[mask][:200]

val_images = train_images[mask][200:250]
val_labels = train_labels[mask][200:250]
```

## (b) Fixed-step Gradient Descent 

```
initial_w = np.random.rand(train_images_subset.shape[1])-0.5
initial_b = np.random.rand()-0.5
```

```
# GD
clf = LogisticRegression(initial_w, initial_b)
loss, valLoss = clf.fit_GD(train_images_subset, train_labels_subset, val_images, val_labels, 20, 5)

print('losses on training set: {}'.format(loss))
print('losses on validation set: {}'.format(valLoss))

predictions_train = clf.predict(train_images_subset)
acc_train = accuracy(predictions_train, train_labels_subset)

predictions_val = clf.predict(val_images)
acc_val = accuracy(predictions_val, val_labels)

print('training set accuracy: {}'.format(acc_train))
print('validation set accuracy: {}'.format(acc_val))
```

## (c) Natural Gradient Descent

```
# NGD
clf = LogisticRegression(initial_w, initial_b)
loss, valLoss = clf.fit_NGD(train_images_subset, train_labels_subset, val_images, val_labels, 5, 0.1)

print('losses on training set: {}'.format(loss))
print('losses on validation set: {}'.format(valLoss))

predictions_train = clf.predict(train_images_subset)
acc_train = accuracy(predictions_train, train_labels_subset)

predictions_val = clf.predict(val_images)
acc_val = accuracy(predictions_val, val_labels)

print('training set accuracy: {}'.format(acc_train))
print('validation set accuracy: {}'.format(acc_val))
```

## (d) Comparison

```
for i in range(5):
  initial_w = np.random.rand(train_images_subset.shape[1])-0.5
  initial_b = np.random.rand()-0.5
  clf = LogisticRegression(initial_w, initial_b)
  loss, _ = clf.fit_GD(train_images_subset, train_labels_subset, val_images, val_labels, 10, 5)
  clf2 = LogisticRegression(initial_w, initial_b)
  loss_NGD, _ = clf.fit_NGD(train_images_subset, train_labels_subset, val_images, val_labels, 10, 0.1)
  print('epoch {}'.format(i+1))
  print('Gradient descent loss: {}'.format(loss))
  print('Natural gradient descent loss: {}'.format(loss_NGD))
  ax1 = plt.subplot(1, 2, 1)
  plt.plot(range(1, 11), loss, 'o')
  plt.title('Gradient Descent')
  plt.subplot(1, 2, 2, sharey=ax1)
  plt.plot(range(1, 11), loss_NGD, 'o')
  plt.title('Natural Gradient Descent')
  plt.tight_layout()
  plt.show()
  print('='*150)
```

Natural gradient descent converged greatly faster than gradient descent, reducing much more loss in one step than what took gradient descent took 10 steps.
